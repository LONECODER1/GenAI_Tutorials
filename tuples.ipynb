```python
# Tuples => Tuples are ordered collections of items that are immutable. They are similar to lists,but their immutability makes them different.


## creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))

lst=list()
print(type(lst))
```

```python
mixed_tuple=(1,"Hello World",3.14,True)
print(mixed_tuple)
```

```python
print(mixed_tuple[1])
print(mixed_tuple[-1])
print(mixed_tuple[1:4])
print(mixed_tuple[::-1])
```

```python
## Tuple Operations
numbers=(1,2,3,4,5,6)
concatenation_tuple=numbers+mixed_tuple
print(concatenation_tuple)
print(mixed_tuple*3)
```
